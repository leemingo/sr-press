```python
import pickle
import pandas as pd
```

```python
with open('team_xP.pkl', 'rb') as f:
    x = pickle.load(f)
```

```python
x = pd.read_csv('team_xP.csv')
```

```python
x['press_success'] = x['xP_total'] + x['xP_difference']
```

```python
x.at[14, 'minutes_played'] = 270
x.at[25, 'minutes_played'] = 360
```

```python
x['xP_per_90'] = x['xP_total'] /  x['minutes_played'] * 90
x['xP_difference_per_90'] = x['xP_difference'] /  x['minutes_played'] * 90
x['press_success_per_90'] = x['press_success'] /  x['minutes_played'] * 90
x['press_action_per_90'] = x['actions'] /  x['minutes_played'] * 90
x['press_success_rate'] = x['press_success'] / x['actions']
```

```python
x.sort_values(by=['press_success_rate'], ascending=False)
```

```python
x.sort_values(by=['xP_difference_per_90'], ascending=False)
```
